```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K_x=np.loadtxt('Evsdisp_X-K.dat')
K_y=np.loadtxt('Evsdisp_Y-K.dat')
K_z=np.loadtxt('Evsdisp_Z-K.dat')
```

```python
K_x
```

```python
fig=plt.figure(figsize=(12,8),linewidth=50)
ax = fig.gca()
#fig, ax=plt.subplots(figsize=(8,6),linewidth=50)
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(6)
plt.grid(linestyle='--')

plt.plot(K_x[:,0],K_x[:,1]-K_x[6,1], lw=6,c='mediumaquamarine',marker="o",ms=15,
         label='$\mathbf{K(x-axis)}$')
# plt.plot(K_y[:,0],K_y[:,1]-K_y[6,1],lw=6,c='orange',marker="o",ms=15,
#          label='$\mathbf{K(y-axis)}$')
plt.plot(K_z[:,0],K_z[:,1]-K_z[6,1], lw=6,c='darkorchid',marker="o",ms=15,
         label='$\mathbf{K(z-axis)}$')


plt.xlabel('Atomic Displacement ($\mathbf{\AA}$)',size=42, fontweight="bold")
plt.ylabel('Energy (eV)',size=42, fontweight="bold")
plt.legend(loc=9, handlelength=2, borderpad=0.5, fontsize=28, shadow=True)

#plt.text(-0.035, 0.0145, '(a) K$\mathbf{_4}$ZnAs$\mathbf{_2}$', fontsize = 42, fontweight="bold")

# plt.xlim(0,800)
#plt.ylim(-0.001,0.0135)
plt.xticks([-0.06,-0.04,-0.02,0.00,0.02,0.04,0.06], fontsize=30,fontweight="bold")
plt.yticks(fontsize=38,fontweight="bold")
plt.tick_params(direction='in', axis='both', length=15, width=6)


fig.tight_layout()
#plt.show()
#plt.savefig('PES_Ag-Ga.png',dpi=500)
```
