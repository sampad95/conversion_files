```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Ga_x=np.loadtxt('Evsad_X-Ga.dat')
Ga_z=np.loadtxt('Evsad_Z-Ga.dat')

Ag_x=np.loadtxt('Evsad_X-Ag.dat')
#Ag_y=np.loadtxt('Evsad_Y-Ag.dat')
Ag_z=np.loadtxt('Evsad_Z-Ag.dat')

Sn_x=np.loadtxt('Evsad_X-Sn.dat')
Sn_z=np.loadtxt('Evsad_Z-Sn.dat')

Se_x=np.loadtxt('Evsad_X-Se.dat')
Se_z=np.loadtxt('Evsad_Z-Se.dat')
```

```python
Ag_x
```

```python
fig=plt.figure(figsize=(12,10),linewidth=50)
ax = fig.gca()
#fig, ax=plt.subplots(figsize=(8,6),linewidth=50)
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(4)
plt.grid(linestyle='--')


plt.plot(Ga_x[:,0],Ga_x[:,1]-Ga_x[6,1], lw=4,c='#0000FF',linestyle='-',
         marker="o",ms=15,label='$\mathbf{Ga(x(y)-axis)}$')
plt.plot(Ga_z[:,0],Ga_z[:,1]-Ga_z[6,1], lw=4,c='#0000FF',linestyle='--',
         marker="o",ms=15,label='$\mathbf{Ga(z-axis)}$')

plt.plot(Ag_x[:,0],Ag_x[:,1]-Ag_x[6,1], lw=4,c='#7F7F7F',linestyle='-',
         marker="o",ms=15,label='$\mathbf{Ag(x(y)-axis)}$')
#plt.plot(Ag_y[:,0],Ag_y[:,1],lw=4,c='orange'
#,marker="o",ms=15,label='$\mathbf{Ag(y-axis)}$')
plt.plot(Ag_z[:,0],Ag_z[:,1]-Ag_z[6,1], lw=4,c='#7F7F7F',linestyle='--',
         marker="o",ms=15,label='$\mathbf{Ag(z-axis)}$')

plt.plot(Sn_x[:,0],Sn_x[:,1]-Sn_x[6,1], lw=4,c='#800080',linestyle='-',
         marker="o",ms=15,label='$\mathbf{Sn(x(y)-axis)}$')
plt.plot(Sn_z[:,0],Sn_z[:,1]-Sn_z[6,1], lw=4,c='#800080',linestyle='--',
         marker="o",ms=15,label='$\mathbf{Sn(z-axis)}$')

plt.plot(Se_x[:,0],Se_x[:,1]-Se_x[6,1], lw=4,c='#FFA500',linestyle='-',
         marker="o",ms=15,label='$\mathbf{Se(x(y)-axis)}$')
plt.plot(Se_z[:,0],Se_z[:,1]-Se_z[6,1], lw=4,c='#FFA500',linestyle='--',
         marker="o",ms=15,label='$\mathbf{Se(z-axis)}$')


plt.xlabel('Atomic Displacement ($\mathbf{\AA}$)',size=28, fontweight="bold")
plt.ylabel('Energy (ev)',size=28, fontweight="bold")

plt.legend(loc=9, handlelength=4, borderpad=0.5, fontsize=20, shadow=True)

#plt.text(400, 1.15, '(b) InAgGeSe$\mathbf{_4}$', fontsize = 28, fontweight="bold")
# plt.text(530, 0.6, '(c)', fontsize = 20, fontweight="bold")
# plt.xlim(0,800)
#plt.xticks([100,200,300,400,500,600,700,800,900,1000], fontsize=20,fontweight="bold")
plt.xticks([-0.06,-0.04,-0.02,0.00,0.02,0.04,0.06], fontsize=28,fontweight="bold")
plt.ylim(-0.005, 0.08)
plt.yticks(fontsize=28,fontweight="bold")
plt.tick_params(direction='in', axis='both', length=10, width=4)


fig.tight_layout()
#plt.show()
#plt.savefig('PES-Ga.png',dpi=500)
```
